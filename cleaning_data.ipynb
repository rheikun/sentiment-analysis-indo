```python
import pandas as pd

import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('twitter_dataset_indo.csv')
print(df.head())
```

```python
# Plotting the sentiment distribution
sentiment_counts = df['sentimen'].value_counts()
sentiment_counts.plot(kind='bar', color=['red', 'blue', 'green'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.show()
```

```python
# Find the minimum count of each sentiment
min_count = sentiment_counts.min()

# Sample the minimum count of each sentiment
df_balanced = df.groupby('sentimen').apply(lambda x: x.sample(min_count)).reset_index(drop=True)

print(df_balanced['sentimen'].value_counts())
```

```python
# Plotting the sentiment distribution
sentiment_counts = df_balanced['sentimen'].value_counts()
sentiment_counts.plot(kind='bar', color=['red', 'blue', 'green'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.show()
```

```python
df_balanced.to_csv('balanced_twitter_dataset_indo.csv', index=False)
```
